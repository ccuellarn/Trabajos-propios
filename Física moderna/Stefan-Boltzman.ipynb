```python
from scipy import constants as C
import numpy as np
import matplotlib.pyplot as plt
```

```python
h=C.h
c=C.c
k=C.k
la=np.linspace(0.000001,0.0003,10000)
def I(l,T):
    a= 2*h*np.pi*(c**2)
    b=(h*c)/(k*T)
    d=a/((l**5)*(np.exp(b/l)-1))
    return d

T9=I(la,90)

plt.plot(la,T9,color='k')
plt.title('T=90K')
plt.xlabel('Longitud de onda')
plt.ylabel('Densidad de potencia')
plt.show()
```

```python
#calibracin de px a nm
def calibracion(px):
    return 0.0024*px

lparalelo =[]
for i in sparalelo:
    lparalelo.append(calibracion(i))

lparalelo
```

```python
#para energias se multiplica
def energiaev(l):
    return ((h*c)/l)*(6.24e18)
    
eparalelo=[]
for i in lparalelo:
    eparalelo.append(energiaev(i))
    
eparalelo
```
