```python
_x = np.linspace(np.min(I1),np.max(I1),2)
param = GetFit(I1,V1)
ym = GetModel(_x,param)

plt.scatter(I1,V1,label = "B=2.5 mT")
plt.plot(_x,ym,color='k',label = "Regresion lineal")
plt.text(-2.3, -0.8e-5, r'$y=2.91\times 10^{-6}x -8.10 \times10^{-8}$')
plt.text(-2.3,-0.9e-5,'R^2= {}'.format(0.99))
plt.ylabel('V (voltios)')
plt.xlabel(r'$I$ (A)')
plt.legend()
m1=param
m1
```
