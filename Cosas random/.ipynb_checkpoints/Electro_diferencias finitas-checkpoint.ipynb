**Problema 4**
Dado que la región presenta forma de semicírculo vale la pena resolver este problema con coordenadas polares. La ecuación de Laplace en coordenadas polares está dada por:
$$ \nabla^2 V(r,\phi) = \frac{\partial^2 V}{\partial r^2} + \frac{1}{r}\frac{\partial V}{\partial r} + \frac{1}{r^2}\frac{\partial^2 V}{\partial \phi^2}= 0 $$
Definiendo los puntos iniciales $(r_0,\phi_0)$ en los nodos internos del potencial de la cuadrícula; con pasos $\Delta r$ y $\Delta \phi$ a través de la misma, cada derivada puede obtenerse de forma discreta al expandir en términos de la serie de Taylor.

Para un desplazamiento en $r\pm h$ con $h$ pequeño:
$$V(r_0+h) = V(r_0)+ \Delta r V'(r_0)+ \frac{(\Delta r)^2}{2}V''(r_0) $$
$$V(r_0-h) = V(r_0)- \Delta r V'(r_0)+ \frac{(\Delta r)^2}{2}V''(r_0) $$
$$\rightarrow V(r_0+h)+V(r_0-h) = 2V(r_0) + (\Delta r)^2V''(r_0) \quad V''(r_0)=\frac{V(r_0+h)+V(r_0-h) - 2V(r_0)}{(\Delta r)^2}$$

Para un desplazamiento en $\phi \pm h$ con $h$ pequeño:
$$V(\phi_0+h) = V(\phi_0)+ \Delta \phi V'(\phi_0)+ \frac{(\Delta \phi)^2}{2}V''(\phi_0) $$
$$V(\phi_0-h) = V(\phi_0)- \Delta \phi V'(\phi_0)+ \frac{(\Delta \phi)^2}{2}V''(\phi_0) $$
$$\rightarrow V(\phi_0+h)+V(\phi_0-h) = 2V(\phi_0) + (\Delta \phi)^2V''(\phi_0) \quad V''(\phi_0)=\frac{V(\phi_0+h)+V(\phi_0-h) - 2V(\phi_0)}{(\Delta \phi)^2}$$

Tomando $r_0=i$, $\phi_0=j$ y $h=1$
Así,
$$ \nabla^2 V(r,\phi) = \frac{V(i+1,j)+V(i-1,j) - 2V(i,j)}{(\Delta r)^2} + \frac{1}{r(i)}\frac{V(i,j)-V(i-1,j)}{\partial r} + \frac{1}{r(i)^2}\frac{V(i,j+1)+V(i,j-1) - 2V(i,j)}{(\Delta \phi)^2} $$
Definiendo $\lambda = \Delta r/ \Delta \phi$ y despejando $V(i,j)$ obtenemos: 
$$ V(i,j)= \bigg(\frac{1}{2 - \Delta r/r(i) + 2(\lambda/r(i))^2} \bigg) \bigg( V(i+1,j) + \bigg(1-\frac{dr}{r(i)}\bigg)V(i-1,j) + \bigg(\bigg(\frac{\lambda}{r(i)}\bigg)^2\bigg)\bigg(V(i,j+1) + V(i,j-1)\bigg) \bigg)$$

A con tinuación se muestra el código con el que se desarrolló el ejercicio a partir de la expresión encontrada.

```python
#Puntos para la red de condiciones iniciales.
Nr = 10
Nphi = 10
r = np.linspace(0.1,10,Nr) #De 0cm a 10cm del radio de alambre
phi = np.linspace(0,np.pi/2,Nphi)
dr = r[1] - r[0]
dphi = phi[1] - phi[0]
lambda_ = dr/dphi 
```

```python
#Condiciones de frontera
def ui(r,phi_):
     if r == 10:
         return 100. #Sobre todo el contorno valga 100V 
     if phi_ == phi[0] or phi_ == phi[-1]:
         return 0. #Justo en el eje x y y valga 0V
     else:
        return 0.
def InitT():
    
    T = np.zeros((Nr,Nphi))
    for i in range(0,len(r)):
        for j in range(0, len(phi)):
            T[i, j] = ui(r[i], phi[j])
    return T

#Malla de condiciones de frontera
T = InitT()
```

```python
def Relajacion(V,r,phi,Nit = int(1e2), omega = 1.5 ,tolerancia = 1e-3):
    itmax = 0
    for it in tqdm(range(Nit)):
        dmax = 0.
        for i in reversed(range(1, len(r) - 1)):
            for j in reversed(range(1,len(phi)-1)):
                #Este condicional se hace con el fin de evitar iterar hacia adelante en la frontera 
                if j == len(r) - 1:
                    b = 1 
                else:
                    b = j+1
                #Expresión que se dedujo al inicio
                c = 2 - dr/r[i] + 2*(lambda_/r[i])**2
                v = V[i+1,j] + (1-dr/r[i])*V[i-1,j] + ((lambda_/r[i])**2)*(V[i,b] + V[i,j-1])
                v /= c
                # Matriz de resta, omega será un factor que ayude a determinar el paso entre iteraciones
                r1 = omega*(v - V[i,j])
                # Actualizo la matrix
                V[i,j] += r1
                if np.abs(r1) > dmax:
                    dmax = r1
        if np.abs(dmax) < tolerancia:
            itmax = it
            break
            
    return T,itmax

T = InitT()
Tf1,_ =  Relajacion(T,r,phi)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
ax1 = fig.add_subplot(122, projection='3d')

R, Phi = np.meshgrid(r,phi)
X,Y = R*np.cos(Phi),R*np.sin(Phi)

ax.contourf(X,Y,T.T,cmap=cm.gnuplot)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Potencial encerrado entre los alambres con 1000 nodos')
ax1.plot_surface(X,Y,T.T,cmap=cm.gnuplot)

plt.show()
```
